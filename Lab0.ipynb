3. Rectangle around messi and the ball\
messi:

```python
#messsi:
start_point = (70,60)
end_point = (460,340)
color = (0,0,255)
thickness = 2
cv2.rectangle(messi_image, start_point, end_point, color, thickness)
plt.imshow(messi_image)
```

ball:

```python
start_point = (340,290)
end_point = (390,335)
color = (0,255,0)
thickness = 2
cv2.rectangle(messi_image, start_point, end_point, color, thickness)
plt.imshow(messi_image)
```

4. add Text

```python
origin=(230,55)
font=2
color=(0,255,0)
thickness=2
font_scale=1
cv2.putText(messi_image,'Messi',origin,font,font_scale,color,thickness,cv2.LINE_AA)
origin=(330,290)
color=(0,0,255)
cv2.putText(messi_image,'Ball',origin,font,font_scale,color,thickness,cv2.LINE_AA)
plt.imshow(messi_image)
```

5. Empty row above and below messi

```python
messi_image = cv2.copyMakeBorder(src=messi_image, top=40, bottom=40, left=0, right=0, borderType=cv2.BORDER_ISOLATED, value=(0, 150, 100))
plt.axis('off')
plt.imshow(messi_image)
```

6. Text in the empty area

```python
origin=(200,30)
font=3
color=(255,0,0)
thickness=2
font_scale=1
cv2.putText(messi_image,'I Like',origin,font,font_scale,color,thickness,cv2.LINE_AA)
font=3
origin=(200,410)
cv2.putText(messi_image,'Trains!!',origin,font,font_scale,color,thickness,cv2.LINE_AA)
plt.imshow(messi_image)
```
